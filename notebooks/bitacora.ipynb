## Bitácora de IA

### Prompt 1: Modelado de CSP y Red Distribuida
**Prompt utilizado:**
"Actúa como un ingeniero de software. Necesito implementar el esqueleto de un problema CSP para una red de servidores en Python. Crea una clase abstracta pura para un CSP (Variables, Dominios, Restricciones). Luego, crea una clase separada Graph que genere nodos y aristas aleatorias simulando una red de computadoras. No escribas todavía los algoritmos de resolución, solo la arquitectura base orientada a objetos."

**Reflexión:**
Estructurar el código de esta forma nos garantizó que el modelo de Factor Graphs quedara estrictamente separado. Validamos que las restricciones binarias se instanciaran correctamente como funciones matemáticas $f_{ij}(X_i, X_j)$ conectando nodos adyacentes del grafo, lo cual respeta al pie de la letra la teoría de variables independientes y factores condicionantes vista en clase.

### Prompt 2: Backtracking y Forward Checking (Lookahead)
**Prompt utilizado:**
"Usando las clases base CSP que construimos, implementa un solucionador genérico de 'Backtracking Search'. Sin embargo, necesito inyectarle una función de inferencia para aplicar 'Forward Checking' (Lookahead). Asegúrate de que, al asignar un valor temporal, esta función pode estrictamente ese valor de los dominios pendientes de los vecinos conectados, y restaure los dominios intactos si es necesario aplicar un backtrack (retroceso)."

**Reflexión:**
Al revisar la recursividad del código generado, comprobamos que el Forward Checking descarta ramas completas exitosamente cumpliendo la regla de que si un sub-dominio $\approx \emptyset$, el factor de probabilidad del sub-árbol completo es cero. La restauración del estado durante el retroceso evita la contaminación de dominios, garantizando la completitud computacional.
