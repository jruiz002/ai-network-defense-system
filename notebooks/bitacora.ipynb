### Prompt 2: Backtracking y Forward Checking (Lookahead)
**Prompt utilizado:**
"Usando las clases base CSP que construimos, implementa un solucionador genérico de 'Backtracking Search'. Sin embargo, necesito inyectarle una función de inferencia para aplicar 'Forward Checking' (Lookahead). Asegúrate de que, al asignar un valor temporal, esta función pode estrictamente ese valor de los dominios pendientes de los vecinos conectados, y restaure los dominios intactos si es necesario aplicar un backtrack (retroceso)."

**Reflexión:**
Al revisar la recursividad del código generado, comprobamos que el Forward Checking descarta ramas completas exitosamente cumpliendo la regla de que si un sub-dominio $\approx \emptyset$, el factor de probabilidad del sub-árbol completo es cero. La restauración del estado durante el retroceso evita la contaminación de dominios, garantizando la completitud computacional.

### Prompt 3: Implementación de Minimax y Poda Alfa-Beta
**Prompt utilizado:**
Implementa en Python Minimax puro y Minimax con Poda Alfa-Beta para un juego de captura de nodos en un grafo. MAX busca maximizar el score, MIN minimizar. Ambos deben recibir profundidad d_max=4, un contador de nodos expandidos, y retornar la mejor acción junto a su valor.

**Reflexión:**
Se verificó que la poda Alfa-Beta no altera la decisión óptima: ambos algoritmos eligieron Server_3 como primera jugada, confirmando que las ramas podadas son irrelevantes. La condición de poda α ≥ β es correcta: en nodos MAX se poda cuando el valor encontrado supera β (MIN nunca lo elegiría), y en nodos MIN cuando cae bajo α (MAX nunca lo elegiría). El contador de nodos confirmó una reducción del 60.5%, consistente con la mejora teórica de O(b^d) a O(b^(d/2)) vista en clase.

### Prompt 4: Función de Evaluación Heurística Eval(s)
**Prompt utilizado:**
Implementa una función heurística Eval(s) para el juego de captura de nodos que estime el valor de un estado sin llegar al nodo terminal. Debe combinar: valor acumulado ya capturado, valor de los nodos en la frontera de expansión de cada jugador, y diferencia de movilidad. Usa pesos 0.5, 0.35 y 0.15 respectivamente.

**Reflexión:**
La función Eval(s) = 0.5·V(s) + 0.35·M(s) + 0.15·F(s) es matemáticamente válida porque V(s) es idéntica al score() del estado terminal, garantizando que si el juego terminara en ese instante el valor sería exacto. Los componentes M y F agregan información prospectiva sin expandir el árbol, aproximando lo que Minimax descubriría a mayor profundidad. Se validó que la función nunca invierte el orden de preferencia entre dos estados donde uno es objetivamente mejor, cumpliendo la propiedad de monotonicidad requerida para que una heurística sea admisible como aproximación del valor Minimax real.

### Prompt 5: Expectiminimax bajo Incertidumbre
**Prompt utilizado:**
"Modifica el motor de juego para incluir incertidumbre: cada acción ahora tiene un 20% de probabilidad de fallar (perder turno). Implementa el algoritmo Expectiminimax agregando 'nodos de azar' entre las decisiones de MAX y MIN. La función debe calcular el valor esperado de cada movimiento ponderando el éxito (0.8) y el fracaso (0.2) en lugar de solo propagar el valor del hijo."

**Reflexión:**
El código implementa correctamente la ecuación $Valor(s) = \sum P(r|s,a) \cdot U(r)$, donde los nodos de azar se ramifican según la distribución de probabilidad definida. Matemáticamente es consistente con la teoría de utilidad esperada de Von Neumann-Morgenstern, ya que el agente maximiza el promedio ponderado de los resultados futuros, penalizando adecuadamente las ramas de alto riesgo frente a las seguras.

### Prompt 6: Análisis Estocástico y Ecuación de Bellman (MDP)
**Prompt utilizado:**
"Genera un script de simulación que enfrente al agente Minimax (determinista) y al Expectiminimax (estocástico) contra un oponente aleatorio para medir tasas de victoria. Además, asumiendo que eliminamos al oponente, formula la Ecuación de Bellman $V(s)$ para este problema tratado como un MDP con costo de movimiento, considerando el factor de descuento y las probabilidades de transición."

**Reflexión:**
El análisis confirmó empíricamente que Expectiminimax es la estrategia racional óptima en entornos ruidosos, sacrificando ganancia máxima potencial por seguridad (100% win-rate). La ecuación formulada $V(s) = \max_a [ 0.8(V_{nodo} - 1 + \gamma V(s')) + 0.2(-1 + \gamma V(s)) ]$ satisface la propiedad de subestructura óptima de la programación dinámica, modelando correctamente el compromiso entre recompensa inmediata y costo por latencia.
